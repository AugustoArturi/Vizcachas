```python
# To support both python 2 and python 3
from __future__ import division, print_function, unicode_literals

# Common imports
import os

# to make this notebook's output stable across runs
np.random.seed(42)

# To plot pretty figures
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

# Where to save the figures
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "Kevin_Feehan"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)


import warnings
warnings.filterwarnings(action="ignore", message="^internal gelsd")
```

```python
plt.figure(figsize=(15,10))
plt.scatter(dfHormonas.Edad,dfHormonas.Progesterona, s=50, c=dfHormonas.Ciclo, alpha=0.4)
plt.xticks(np.arange(0,160,5))
plt.title("Valores Progesterona x Edad",fontsize=14)
plt.xlabel("Edad",fontsize=14)
plt.ylabel("Progesterona",fontsize=14)
z = np.polyfit(dfHormonas.Edad,dfHormonas.Progesterona, 1) 
p = np.poly1d(z) 
plt.plot(dfHormonas.Edad,p(dfHormonas.Edad),"r--") 
plt.show()

```

```python
plt.figure(figsize=(15,10))
plt.scatter(dfHormonas.Peso_Kg,dfHormonas.Edad, s=50, c="g", alpha=0.4)

plt.title("Valores Peso_Kg x Edad",fontsize=14)
plt.xlabel("Peso (kg)",fontsize=14)
plt.ylabel("Edad de gestacion",fontsize=14)
plt.show()
save_fig("PesoVsEdad")
```

```python
dfClima=dfClima.loc[(dfClima.Mes!="Septiembre")]
dfClima
```

```python
dfClima=dfClima.loc[(dfClima.Mes!="Marzo")]
dfClima
```

```python
NombreANumero = {'Enero':1, 'Febrero':2, 'Abril':4, 'Mayo':5,"Junio":6,"Julio":7, "Agosto":8,"Septiembre":9,
            "Octubre":10, "Noviembre":11, "Diciembre":12}

dfClima["Mes"] = dfClima.Mes.map(NombreANumero)
```

```python
dfCiclo = pd.merge(dfCiclo, dfClima, on=["Mes","Año"])
dfCiclo
```

```python


data = [
    go.Parcoords(
        line = dict(color = dfCiclo['Ciclo'],
                   colorscale = [[0,'#FF0040'],[1,'#64FE2E']]),
        dimensions = list([
            dict(range = [1,12],
                constraintrange = [4,8],
                label = 'Mes', values = dfCiclo['Mes']),
            dict(range = [2006,2017],
                label = 'Año', values = dfCiclo['Año']),
            dict(range = [8,30],
                label = 'Temp.Media', values = dfCiclo['TMed']),
            dict(range = [13,31],
                label = 'Temp.Maxima', values = dfCiclo['TMax']),
             dict(range = [0,20],
                label = 'Temp. Min', values = dfCiclo['TMin']),
             dict(range = [52,90],
                label = 'Humedad', values = dfCiclo['Humedad']),
             dict(range = [6,220],
                label = 'Precipitacion', values = dfCiclo['Precipitacion']),
             dict(range = [8,20],
                label = 'Viento Medio', values = dfCiclo['VientoMed']),
            dict(range = [15,30],
                label = 'Viento Max', values = dfCiclo['VientoMax']),
              dict(range = [7,10],
                label = 'Visibilidad', values = dfCiclo['Visibilidad']),
        ])
    )
]

layout = go.Layout(
    plot_bgcolor = '#E5E5E5',
    paper_bgcolor = '#E5E5E5'
)

fig = go.Figure(data = data, layout = layout)
py.iplot(fig, filename = 'parcoords-basic')
```
